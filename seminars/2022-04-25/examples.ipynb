```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_start = -10
x_end = 10

plt.figure(figsize=(8, 6), dpi=80)
plt.subplot(1, 1, 1)
X = np.linspace(x_start, x_end, 200)
C, S = np.cos(X), np.sin(X)

plt.plot(X, C, color="red", linewidth=10, linestyle=":")
plt.plot(X, S, color="green", linewidth=1, linestyle="-")

plt.xlim(x_start * 1.1, x_end * 1.1)
plt.ylim(C.min() * 1.1, C.max() * 1.1)
plt.xticks(np.linspace(-4, 4, 9))
plt.yticks(np.linspace(-1, 1, 7))
plt.show()
```

```python
x_start = -30
x_end = 30
X = np.linspace(x_start, x_end, 200)

# y = ax^2 + bx + c: 3 набора параметров
a, b, c = 1, 0, 0  # y = x^2
a2, b2, c2 = -1, -10, 2  # y = -x^2 - x + 2
a3, b3, c3 = 2, 10, -10  # y = 2x^2 + 10x - 10

Y1 = np.array([a*x**2 + b*x + c for x in X])
Y2 = np.array([a2*x**2 + b2*x + c2 for x in X])
Y3 = np.array([a3*x**2 + b3*x + c3 for x in X])
plt.plot(X, Y1, linewidth=1, linestyle="-")
plt.plot(X, Y2, linewidth=1, linestyle="-")
plt.plot(X, Y3, linewidth=1, linestyle="-")
plt.xlim(x_start * 1.1, x_end * 1.1)
plt.ylim(min(Y1.min(), Y2.min(), Y3.min()), max(Y1.max(), Y2.max(), Y3.max()))
plt.show()
```

```python
t = 2 * np.pi / 3


#plt.plot([t, t], [0, np.cos(t)], color='blue', linewidth=2.5, linestyle="--")
#plt.scatter([t, ], [np.cos(t), ], 50, color='blue')
#plt.scatter([t, ],[np.sin(t), ], 50, color='red')
#plt.plot([t, t],[0, np.sin(t)], color='red', linewidth=2.5, linestyle="--")

plt.annotate(r'$cos(\frac{2\pi}{3})=-\frac{1}{2}$',
             xy=(t, np.cos(t)), xycoords='data',
             xytext=(90, -50), textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))


plt.annotate(r'$\sin(\frac{2}{3})$',
             xy=(t, np.sin(t)), xycoords='data',
             xytext=(+10, +30), textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=10"))
plt.xlim(-100, 100)
plt.ylim(-100, 100)
```

```python
def make_plot(number, X, Y):
    plt.subplot(2, 2, number)
    plt.xlim(0, 1)
    plt.ylim(0, 1)
    plt.plot(X, Y)
    
fig = plt.figure()
make_plot(1, (0.5, 0.5, 1), (0, 0.5, 0.5))
make_plot(2, (0, 0.5, 0.5), (0.5, 0.5, 0))
make_plot(3, (0.5, 0.5, 1), (1, 0.5, 0.5))
make_plot(4, (0, 0.5, 0.5), (0.5, 0.5, 1))
plt.show()
```

```python
fig = plt.figure()
X = np.linspace(0, 9, 300)
Y = np.tan(X)

for num in range(6):
    plt.subplot(2, 3, num + 1)
    x = num % 3 * 3
    plt.xlim(x, x + 3)
    y = -(num // 3 * 10)
    plt.ylim(y, y + 10)    
    plt.plot(X, Y)

plt.show()
```
